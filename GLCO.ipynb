```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200 # 200 e.g. is really fine, but slower
```

```python
df = pd.read_csv('GLCO.csv')
df = df.drop(['Open', 'Close', 'Adj Close'], axis=1)
df
```

```python
xs = np.arange(len(df.High))

plt.yscale('log')
plt.plot(xs, df.High)
plt.plot(xs, df.Volume)

plt.show()
```
