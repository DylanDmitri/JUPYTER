```python
from itertools import product

nums = tuple(range(1,MAXVAL))

results = []

for i,outcome in enumerate(product(*[nums]*rolls)):
    if min(outcome) == max(outcome):
        continue
    results.append(min(outcome))

np.mean(results)
```
