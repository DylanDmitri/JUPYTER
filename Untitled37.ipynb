```python
import numpy as np


class ParseState:
    cast = 'cast'

class EachVector(np.ndarray):
    def __new__(cls, values):
        self = np.asarray(list(values)).view(cls)
        self._fsm = None
        return self
    
    def __getattr__(self, key):
        if key == 'to':
            self._fsm = ParseState.cast
            return self

        if self._fsm == ParseState.cast:
            self._fsm = None
            newtype = eval(key)
            return EachVector(map(newtype, self))

        # ===== regular operation
        value = getattr(self[0], key)
        
        if value is None:
            raise Exception("Field not found.")

        elif callable(value):
            return lambda *args, **kwargs: EachVector(
                getattr(e, key)(*args, **kwargs)
                for e in self
            )
        
        else:
            return EachVector(getattr(e, key) for e in self)


def each(col):
    return EachVector(col)

df = pd.read_csv('chipotle.tsv', delimiter='\t')
df
```
